```python
import pandas as pd;
import matplotlib.pyplot as plt;
import numpy as np;
```

```python
cities_df = pd.read_csv('./data/cities.csv')
global_df = pd.read_csv('./data/global.csv')
```

```python
guarulhos_df = cities_df.query('city == "Guarulhos"').sort_values('year').reset_index(drop=True)
guarulhos_df.head()
```

```python
global_df.head()
```

```python
guarulhos_df['m7yr_average'] = guarulhos_df.iloc[:,3].rolling(7).mean()
guarulhos_df['m12yr_average'] = guarulhos_df.iloc[:,3].rolling(12).mean()
```

```python
plt.figure(figsize=(20,10));
plt.plot(guarulhos_df['year'],guarulhos_df['m7yr_average']);
plt.plot(guarulhos_df['year'],guarulhos_df['m12yr_average']);
plt.xlabel('Year')
plt.ylabel('Temperature Average [ºC]')
plt.title('Local temperature average')
plt.grid(alpha=0.3)
plt.legend(['7 Years Move Average','12 Years Move Average'])
```

```python
global_df['m7yr_average'] = global_df.iloc[:,1].rolling(7).mean()
global_df['m12yr_average'] = global_df.iloc[:,1].rolling(12).mean()
```

```python
guarulhos_df.query('year > 1860')
```

```python
plt.figure(figsize=(20,10));
plt.plot(guarulhos_df['year'],guarulhos_df['m12yr_average']);
plt.plot(global_df['year'],global_df['m12yr_average']);
plt.xlim(1865,2015)
plt.xlabel('Year')
plt.xticks(np.arange(1865,2020,step=10))
plt.yticks(np.arange(0,25,step=5))
plt.ylabel('Temperature Average [ºC]')
plt.title('Annual average temperature data')
plt.grid(alpha=0.3)
plt.legend(['Global data','Local data']);
```
